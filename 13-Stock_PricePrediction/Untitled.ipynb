# Stock Price Prediction
Cody Has collected the stock prices of coding blocks for past 5 year from 2012 to 2016, Now using this data he wants to predict the Stock Prices for the year 2017 so he can decide if investing in the stocks will be profitable or not. Data Description:

Data consists of two file Train and Test.

Train File has two csv files xtrain and ytrain, xtrain consists the high of stock prices of the current day and ytrain consists of high of stock price for the next day.

Now by training your model on this data try to predict the stock prices of next 20 days in 2017 using x_test.

The submission file format should be same as y_train

```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import *
```

```python
X_train = pd.read_csv('x_train.csv').values
X_test  = pd.read_csv('x_test.csv').values
Y_train = pd.read_csv('y_train.csv').values
```

```python
u=X_train.mean()
s=X_train.std()

X_train=(X_train-u)/s
X_test=(X_test-u)/s
```

```python
model = Sequential()
model.add(LSTM(30,activation="relu",input_shape=(1,1),return_sequences=True))
model.add(LSTM(30,activation="relu"))
model.add(Dense(20,activation="relu"))
model.add(Dense(1))
model.summary()
```

```python
X=X_train.reshape((-1,1,1))
model.compile(loss='mean_squared_error', optimizer='adam')
hist=model.fit(X, Y_train,validation_split=0.1,epochs=1000, batch_size=128)
```

```python
Y_pred=model.predict(X_test.reshape((-1,1,1)))
```

```python
Y_pred1=Y_pred.reshape((-1,1))
```
