You will be given an input image and two image templates(one for sunglasses & moustache , you task is to overlay the eyeglasses and moustache on the given character. A sample input output is show below. To detect the facial keypoints like position of eyes and nose, we also provide you with

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
temp=cv2.imread("Before.png")
temp=cv2.cvtColor(temp,cv2.COLOR_BGR2RGB)
```

```python
main=cv2.imread("Before.png")
main=cv2.cvtColor(main,cv2.COLOR_BGR2RGB)
```

```python
glass=cv2.imread("glasses.png",-1)
mustache=cv2.imread("mustache.png",-1)
```

```python
eye_casscade=cv2.CascadeClassifier("C:\\Users\\User\\Desktop\\ML CB\\Day3\\frontalEyes35x16.xml")
eyes=eye_casscade.detectMultiScale(temp, 1.1, 10)
```

```python
for x,y,w,h in eyes:
    print(x,y,w,h)
    glasses=cv2.resize(glass,(h,w))
    for i in range(w):
        for j in range(h):
            if(glasses[i][j][-1]>0):
                temp[y+i,x+j,:]=glasses[i,j,:-1]
    #cv2.rectangle(temp,(x,y),(x+h,y+w),(0,0,0),2)
plt.figure(figsize=(6,6))
plt.imshow(temp)
plt.show()
```

```python
nose_casscade=cv2.CascadeClassifier("C:\\Users\\User\\Desktop\\ML CB\\Day3\\Nose18x15.xml")
noses=nose_casscade.detectMultiScale(main, 1.1, 3)
```

```python
for x,y,w,h in noses:
    mustaches=cv2.resize(mustache,(h,w))
    k=int(.5*w)
    for i in range(w):
        for j in range(h):
            if(mustaches[i][j][-1]>0):
                temp[y+k+i][x+j][:]=mustaches[i][j][:-1]
    #cv2.rectangle(temp,(x,y),(x+h,y+w),(0,0,0),2)
plt.figure(1,(10,10))
plt.imshow(temp)
plt.show()
```
